```python
# !pip install requests
# !pip install beautifulsoup4

import requests
from bs4 import BeautifulSoup
import csv
import random
from tkinter import *
from tkinter.ttk import *

words = []
variables = []

with open('final_EnglishGame_dataset.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for row in csv_reader:
      words.append(row[0])

random.shuffle(words)
words = words[:5]
meanings = []
word_meaning = {}
words = ["cat","bat","mat","sit","rat"]

for word in words:
  web = requests.get('https://www.lexico.com/en/definition/'+word)
  # print(word,":")
  data = web.content
  soup = BeautifulSoup(data,features="html.parser")
  # print(soup)
  tag = soup.find_all("span","ind")
  # print(tag)

  for i in tag:
    text = i.text
    meanings.append(text)
    word_meaning[word]=text
    break

user_ans={}
random.shuffle(words)
random.shuffle(meanings)

ans=[]
def retrieve_meaning():
    for i in ans:
        print(i)
    for i in variables:
        print(i)
        
    for i in range(len(words)):
      user_ans[words[i]] = meanings[ans-1]

    score = 0
    for i in list(word_meaning.keys()):
      if(user_ans[i]==word_meaning[i]):
        score+=1


    # print(words)
    # print(meanings)
    # print(word_meaning)
    # print(user_ans)
    print("\nScore: ",score,"/",len(user_ans))

    print("\nCorrect answers are - ")
    for i in list(word_meaning.keys()):
      print(i,"â–º",word_meaning[i],"\n")

def meaning():
    mean = Toplevel()
    mean.title('Word Play')
    mean.resizable(width=False, height=False)
    mean.geometry("400x780")
    
    bg1 = PhotoImage(file = "temp_background.gif")
    l = Label( mean, image = bg1)
    l.place(x = 0, y = 0)
    
    frame = Frame(mean,width=350,height=500)
    frame.pack(pady=20,padx=20)
    frame.place(x = 20, y = 50)
    
    head = Label(frame, text ="Match the given words with there meaning - ",font=("Arial", 11))
    head.config(anchor=CENTER)
    head.pack(fill=BOTH, expand=1,pady=10)

    subhead = Label(frame, text =f'{"Words":{20}} {" ":{5}} Meanings(Options)',font=("Arial", 10))
    subhead.config(anchor=CENTER)
    subhead.pack(fill=BOTH, expand=1,pady=5)

    for i in range(len(words)):
        ques = Label(frame, text =f'{str(chr(65+i))+") "+words[i]:{20}} {"-":{5}} {str(i+1)+") "+meanings[i]}',font=("Arial", 10),wraplength=370)
        ques.config(anchor=CENTER)
        ques.pack(fill=BOTH, expand=1,pady=5)

    ans_head = Label(frame, text ="Your answers here - ",font=("Arial", 11),wraplength=350)
    ans_head.config(anchor=CENTER)
    ans_head.pack(fill=BOTH, expand=1,pady=10)
    
    a=0
    

    for i in range(len(words)):
      va = StringVar()
      q = Label(frame, text=f'{str(chr(65+i))+") "+words[i]:{20}} {"-":{5}}',font=("Arial", 10)).pack()
      a = Entry(frame,textvariable=va)
      a.pack()
      variables.insert(i,va)
      ans.insert(i,a)
      
      
    print(ans)
    print(variables)
    
    button = Button(frame, text = "Submit", command = retrieve_meaning)
    button.pack(padx = 5, pady = 5)
    
    mean.mainloop()

# meaning()
```
